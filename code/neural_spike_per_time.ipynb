```python
# for mouse2
# find indices of each cell t
v_index = np.where(type2 > 1.5)[0]
#d_index = np.where(0.5>type2>1.5)
m_index = np.where(type2<1)[0]
```

```python
# now plot the times...

# select the ventral neurons
activity = [] #activity per neuron initialized from time t
for i in range(len(reach2)): # for each reach
    spike= reach2[i][7][v_index]
    #rtime = reach2[i][5][v_index]
    activity.append(spike.sum(axis=0))
```

```python
# select the motor neurons
mactivity = [] #activity per neuron initialized from time t
for i in range(len(reach2)): # for each reach
    spike= reach2[i][7][m_index]
    #rtime = reach2[i][5][v_index]
    mactivity.append(spike.sum(axis=0))
    
```

```python
len(activity[1])
```

```python
activity = np.array(activity)
final = np.average(activity, axis=0)

mactivity = np.array(mactivity)
mfinal = np.average(mactivity, axis=0)
```

```python
%pylab inline
pylab.rcParams['figure.figsize'] = (15, 8)
plt.plot(final*3/2, label='ventral*3/2')
plt.plot(mfinal, label='motor')

plt.title("Neuron activity over time of reach")
plt.xlabel("relative time (ms)")
plt.ylabel("Neuron activity")
plt.legend()
plt.grid()
plt.show()
```
